## Second class - Using sheets

```python
# Getting a sheet link
https://docs.google.com/spreadsheets/d/1pvBoLyX8kP0TjtUbadVMGdTl4yzm6bHMThhPiqCVtpw/edit?usp=sharing
```

```python
# Getting the sheet 'id'
sheet_id = '1pvBoLyX8kP0TjtUbadVMGdTl4yzm6bHMThhPiqCVtpw'
```

```python
# Preparing a url for the sheet
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
# gviz/tq => parameter that allows the access of colab to google sheets through a google API
# tqx=out:csv => returns a .csv file
# &sheet => return all the data inside the sheet
```

```python
# Reading the dataframe
dados_sheet = pd.read_csv(url)
dados_sheet.head()
```

```python
# Writing the dataframe to csv
dados_sheet.to_csv('dados_emissoes_co2.csv', index=False)
```

## Third class - Manipiluating .json files

```python
# Importing libraries to get an API
import requests
import json
```

```python
# Getting the data from the API
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
resultado = json.loads(dados_frutas.text)
```

```python
# Importing Pandas
import pandas as pd

# Creating and normalizing the fruit dataset
df_fruit = pd.json_normalize(resultado)
df_fruit.head()
```

## Third class - Manipulating .json files

```python
# Reading a non-nested json file
dados_pacientes = pd.read_json('/content/pacientes.json')
dados_pacientes.head()
```

```python
# Reading a nested json file
dados_pacientes_2 = pd.read_json('/content/pacientes_2.json')
dados_pacientes_2.head()
```

```python
# Treating the nested json file
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
df_normalizado.head()
```

```python
'''Importante: Na aula nós realizamos a normalização em um arquivo com o formato JSON.
No entanto, a função json_normalize() aceita apenas um dicionário ou uma lista de dicionários.
Por isso, no vídeo foi usada a estratégia de usar o código: pd.json_normalize(dados_pacientes_2['Pacientes']).
Porém, se tentarmos usar parâmetros da função json_normalize em um arquivo JSON podem surgir erros.
Para contornar isso, precisamos importar o módulo json e ler os arquivos conforme o código abaixo:'''

#Importando a biblioteca Pandas
import pandas as pd

#Importando o módulo JSON
import json

#Lendo o arquivo json usando o módulo Python JSON
with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())

#Normalizando os dados com os parâmetros record_path e meta
pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
```

```python
# Writing a dataframe to .json
df_normalizado.to_json('historico_pacientes_normalizado.json')
```

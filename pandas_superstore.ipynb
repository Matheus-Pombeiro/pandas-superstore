## First class - Reading .csv files

```python
# Importing Pandas
import pandas as pd

# Reading the data
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/superstore_data.csv'
dados_mercado = pd.read_csv(url)

# Showing the data
dados_mercado.head()
```

```python
# Readng the data formatted with semicolons
url_ponto_virgula = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/superstore_data_ponto_virgula.csv'
dados_ponto_virgula = pd.read_csv(url_ponto_virgula, sep=';')

# Showing the data
dados_ponto_virgula.head()
```

```python
# Reading the first rows of the dataset
dado_primeiras_linhas = pd.read_csv(url, nrows=5)

# Showing the dataframe
dado_primeiras_linhas
```

```python
# Reading and selecting data
dados_selecao = pd.read_csv(url, usecols=['Id', 'Year_Birth', 'Income'], nrows=5)

# Showing the dataframe
dados_selecao
```

```python
# Reading and selecting data
dados_selecao = pd.read_csv(url, usecols=[0, 1, 4], nrows=5)

# Showing the dataframe
dados_selecao
```

```python
# Writing data in .csv
dados_selecao.to_csv('clientes_mercado.csv', index=False)
# index=False => doesn't duplicate the 'id' column
```

## Second class - Using sheets

```python
# Reding the dataset from a sheet
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'
# ?raw=True => it's used to read the raw file
dados_co2 = pd.read_excel(url)
```

```python
# Showing the data
dados_co2.head()
```

```python
# Verifying the pages in the sheet
pd.ExcelFile(url).sheet_names
```

```python
# Getting the data from the 2nd page of the dataset
percapita = pd.read_excel(url, sheet_name='emissoes_percapita')
percapita.head()
```

```python
# Getting the data from the 3rd page of the dataset
fontes = pd.read_excel(url, sheet_name='fontes')
fontes.head()
```

```python
# Getting the data by an interval of columns
intervalo = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D')
intervalo.head()
```

```python
# Getting the data by an interval of columns and the first ten rows
intervalo_2 = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D', nrows=10)
intervalo_2
```

```python
# Writing a dataframe in and excel file
percapita.to_excel('co2_percapita.xlsx', index=False)
```
